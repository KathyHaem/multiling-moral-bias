```python
import statistics

def calc_score(frame, col_name, numbers, reverse_coded):
  scores = frame[col_name].loc[numbers]
  absolute = [sign * score for sign, score in zip(reverse_coded, scores)]
  mean = statistics.mean(absolute)
  return mean
```

# Bar Plot Definition

```python
import matplotlib.pyplot as plt
plt.rc('font', size=10) #controls default text size


def bar_plot(ax, data, xticks, legend_names, colors=None, total_width=0.8, 
             single_width=1, legend=True, min_y=None, max_y=None):
    """Draws a bar plot with multiple bars per data point.

    Parameters
    ----------
    ax : matplotlib.pyplot.axis
        The axis we want to draw our plot on.

    data: dictionary
        A dictionary containing the data we want to plot. Keys are the names of the
        data, the items is a list of the values.

        Example:
        data = {
            "x":[1,2,3],
            "y":[1,2,3],
            "z":[1,2,3],
        }

    colors : array-like, optional
        A list of colors which are used for the bars. If None, the colors
        will be the standard matplotlib color cyle. (default: None)

    total_width : float, optional, default: 0.8
        The width of a bar group. 0.8 means that 80% of the x-axis is covered
        by bars and 20% will be spaces between the bars.

    single_width: float, optional, default: 1
        The relative width of a single bar within a group. 1 means the bars
        will touch eachother within a group, values less than 1 will make
        these bars thinner.

    legend: bool, optional, default: True
        If this is set to true, a legend will be added to the axis.
    """

    # Check if colors were provided, otherwise use the default color cycle
    if colors is None:
        colors = ["#000000","#004949","#009292","#ff6db6","#ffb6db", "#490092","#006ddb",
                  "#b66dff","#6db6ff","#b6dbff", "#920000","#924900","#db6d00","#24ff24","#ffff6d"]
        colors = ["#d55e00", "#cc79a7", "#0072b2", "#f0e442", "#009e73"]
        # ar cs de en zh
        colors = ["#d73027", "#fc8d59", "#fee090", "#91bfdb", "#4575b4"]
        # colors = ['#4477AA', '#EE6677', '#228833', '#CCBB44', '#66CCEE', '#AA3377', '#BBBBBB']
        # colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

    # Number of bars per group
    n_bars = len(data)

    # The width of a single bar
    bar_width = total_width / n_bars

    if min_y is not None or max_y is not None:
      ax.set_ylim(bottom=min_y, top=max_y)
    # List containing handles for the drawn bars, used for the legend
    bars = []
    # Iterate over all data
    for i, (name, values) in enumerate(data.items()):
        # The offset in x direction of that bar
        x_offset = (i - n_bars / 2) * bar_width + bar_width / 2

        # Draw a bar for every value of that type
        for x, y in enumerate(values):
            bar = ax.bar(x + x_offset, y, width=bar_width * single_width, 
                         #edgecolor="lightgray", 
                         color=colors[i % len(colors)], 
                         label=legend_names[i])

        # Add a handle to the last drawn bar, which we'll need for the legend
        bars.append(bar[0])

    # Draw legend if we need
    if legend:
        ax.legend(bars, legend_names)
    # ax.xticks(range(len(xticks)), xticks)
    ax.set_xticks(range(len(xticks)))
    ax.set_xticklabels(xticks)
```

```python
joeckel_12_german = pd.read_spss('Dataset_MFT_Age_Gender_Country_NDB_SJ_LD_2012_German_only.sav')
joeckel_various = pd.read_spss('Dataset_MFT_Age_Gender_Country_SJ_NDB_LD_various.sav')
joeckel_16 = pd.read_spss('Dataset_MFT_Age_Gender_SJHF2016.sav')
```

```python
def stats(frame):
  """ Taking every mean minus 1 to get scale [0,5] instead of [1,6] """
  print("\t\tmean \tstd")
  print("------------------------------")
  print("Harm \t\t" + str(round(frame.MFQ_HARM.mean() - 1, 2)) + "\t" + str(round(frame.MFQ_HARM.std(), 2)))
  print("Fairness \t" + str(round(frame.MFQ_FAIRNESS.mean() - 1, 2)) + "\t" + str(round(frame.MFQ_FAIRNESS.std(), 2)))
  print("Ingroup \t" + str(round(frame.MFQ_INGROUP.mean() - 1, 2)) + "\t" + str(round(frame.MFQ_INGROUP.std(), 2)))
  print("Authority \t" + str(round(frame.MFQ_AUTHORITY.mean() - 1 , 2)) + "\t" + str(round(frame.MFQ_AUTHORITY.std(), 2)))
  print("Purity \t\t" + str(round(frame.MFQ_PURITY.mean() - 1, 2)) + "\t" + str(round(frame.MFQ_PURITY.std(), 2)))
  print("\n")
```

```python
def spss_bar_plot(ax, frames, names):
  data = {}
  for frame, name in zip (frames, names):
    data[name] = [frame.MFQ_HARM.mean() - 1, frame.MFQ_FAIRNESS.mean() - 1, frame.MFQ_INGROUP.mean() - 1, frame.MFQ_AUTHORITY.mean() - 1, frame.MFQ_PURITY.mean() - 1]
  bar_plot(ax, data, xticks=["Care", "Fairness", "Loyalty", "Authority", "Purity"], legend_names=names, max_y=5)
```

```python
fig, ax = plt.subplots()
spss_bar_plot(ax, 
         [joeckel_12_german,
          joeckel_various,
          joeckel_16],
         ["12 DE", "12 gem", "16 DE"])
```

```python
fig, ax = plt.subplots()
sample = joeckel_12_german.sample(5)
spss_bar_plot(ax,
              [sample.iloc[0],
               sample.iloc[1],
               sample.iloc[2],
               sample.iloc[3],
               sample.iloc[4]], 
              names=["0","1","2","3","4"])
fig.tight_layout()
fig.savefig("human-var-example.pdf")
plt.show()
```
